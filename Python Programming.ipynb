**Inheritance**

* Single Inheritance
* Multiple Inheritance
* Multilevel Inheritance
* Hierarchical Inheritance
* Hybrid Inheritance


```
# Single inheritance ------------------

class Animal:
    def __init__(self):
      print('constructor at parent level')
    def speak(self):
        print("Animal Speaking")
#child class Dog inherits the base class Animal
class Dog(Animal):
    def __init__(self):
      print('constructor at child level')
      super().__init__()
    def bark(self):
        print("dog barking")
d = Dog()
d.bark()
d.speak()
```

```
# Multi - level inheritance -----------------------

class Animal:
    def speak(self):
        print("Animal Speaking")
#The child class Dog inherits the base class Animal
class Dog(Animal):
    def bark(self):
        print("dog barking")
#The child class Dogchild inherits another child class Dog
class DogChild(Dog):
    def eat(self):
        print("Eating bread...")
d = DogChild()
d.bark()
d.speak()
d.eat()
```

```
# Multiple inheritance -----------------------

class Calculation1:
    def Summation(self,a,b):
        return a+b
    def disp(self):
      print('Calculation1')
class Calculation2:
    def Multiplication(self,a,b):
        return a*b
    def disp(self):
      print('Calculation2')
class Derived(Calculation1,Calculation2):
    def Divide(self,a,b):
        return a/b
d = Derived()
# Better way
#c1=Calculation1()
#c2=Calculation2()
#c2.disp()
print(d.Summation(10,20))
print(d.Multiplication(10,20))
print(d.Divide(10,20))
d.disp()
```

Now, let's use `requests` to fetch the content of a webpage.

```
# Import the requests library
import requests

# Define the URL of the webpage to fetch
url = 'http://example.com' # You can replace this with any URL

# Use a try-except block to handle potential errors during the request
try:
    # Make a GET request to the URL
    response = requests.get(url)
    # Raise an exception for bad status codes (e.g., 404 Not Found, 500 Internal Server Error)
    response.raise_for_status()
    # Get the HTML content from the response
    html_content = response.text
    # Print a success message
    print("Successfully fetched the webpage content!")
    # Print the first 500 characters of the HTML content
    print(html_content[:500])
except requests.exceptions.RequestException as e:
    # If an error occurs during the request, print an error message
    print(f"Error fetching the webpage: {e}")
```

Next, we'll use `BeautifulSoup` to parse the HTML content and extract information.

```
# Import the BeautifulSoup class from the bs4 library
from bs4 import BeautifulSoup

# Check if the html_content variable exists and is not empty
if 'html_content' in locals(): # Check if html_content was successfully fetched
    # Create a BeautifulSoup object by parsing the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Example 1: Get the title of the page
    # The title is found within the <title> tag
    title = soup.title.string
    print(f"\nPage Title: {title}")

    # Example 2: Find all the paragraph tags
    # find_all() returns a list of all elements matching the tag
    paragraphs = soup.find_all('p')
    print(f"\nNumber of paragraph tags: {len(paragraphs)}")
    # If there are any paragraphs, print the text of the first one
    if paragraphs:
        print("First paragraph:")
        print(paragraphs[0].text)
else:
    # If html_content was not fetched, print an error message
    print("Could not parse HTML as the webpage content was not fetched.")
```

Here are some more examples demonstrating how to use `BeautifulSoup` to parse HTML and extract data:

```
# Import necessary libraries
from bs4 import BeautifulSoup
import requests

# Fetching a more complex page for demonstration
url = 'http://quotes.toscrape.com/' # A website specifically designed for scraping examples
# Use a try-except block to handle potential errors
try:
    # Make a GET request to the URL
    response = requests.get(url)
    # Raise an exception for bad status codes
    response.raise_for_status()
    # Get the HTML content
    html_content = response.text
    # Create a BeautifulSoup object to parse the HTML
    soup = BeautifulSoup(html_content, 'html.parser')

    print(f"Successfully fetched and parsed: {url}")

    # Example 1: Finding the first element of a specific tag
    print("\n--- Example 1: Finding the first <div> tag ---")
    # find() returns the first element that matches the tag
    first_div = soup.find('div')
    # Print a snippet of the found div using prettify() for formatted output
    print(first_div.prettify()[:200] + "...") # Print a snippet

    # Example 2: Finding all <span> tags with class 'text' (the quotes)
    print("\n--- Example 2: Finding all <span> tags with class 'text' (the quotes) ---")
    # find_all() returns a list of all elements that match the tag and attributes
    quotes = soup.find_all('span', class_='text')
    print(f"Number of quotes found: {len(quotes)}")
    # If quotes are found, print the text of the first 3
    if quotes:
        print("First 3 quotes:")
        for i, quote in enumerate(quotes[:3]):
            print(f"{i+1}. {quote.text}")

    # Example 3: Navigating the parse tree (finding the author of a quote)
    print("\n--- Example 3: Finding the author of the first quote ---")
    # Find the first quote element
    first_quote_element = soup.find('span', class_='text')
    if first_quote_element:
        # The author is in a <small> tag within the same parent <div>, find the element with class 'author'
        # The author is within the same parent div as the quote, but not necessarily a sibling.
        # We can find the parent div first, and then search for the author within that div.
        parent_div = first_quote_element.find_parent('div', class_='quote')
        if parent_div:
            author_element = parent_div.find('small', class_='author')
            if author_element:
                print(f"Author of the first quote: {author_element.text}")
            else:
                print("Author element not found within the parent div.")
        else:
            print("Parent div with class 'quote' not found for the first quote.")
    else:
        print("First quote element not found.")


except requests.exceptions.RequestException as e:
    # Handle errors during the request
    print(f"Error fetching the webpage: {e}")
except Exception as e:
    # Handle any other errors during parsing
    print(f"An error occurred during parsing: {e}")
```
